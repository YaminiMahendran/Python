```python
#Identifying the users
users=[]
for element in list1:
    di = element.get('statuses')
    for tweet in di:
        users.append(tweet['user']['name'])
print(users)
```

```python
#Displaying the locations where the tweets were sent
locations=[]
for element in list1:
    di = element.get('statuses')
    for tweet in di:
        locations.append(tweet['user'])
for each in locations:
    print(each['location'])
```

```python
#Displaying all the tweets
tweets=[]
for element in list1:
    di = element.get('statuses')
    for tweet in di:
        tweets.append(tweet['text'])
for each in tweets:
    print(each)
```

```python
#Displaying all hashtags and getting the count of hashtags
hashtag =[]
for element in list1:
    di = element.get('statuses')
    for each in di:
        all_hashtags=each['entities']['hashtags']
        for each in all_hashtags:
            hashtag.append(each['text'])
print(hashtag)
count={}
for word in hashtag:
    count[word] = count.get(word,0) + 1
print(count)
      
```

```python
#Displaying Top 5 hashtags
import collections
count = collections.Counter(count)
count.most_common(5)
```

```python
#removing stopwords,punctuations
import nltk
import string
nltk.download('stopwords')
word_tweets=[]
for each in tweets:
    word_tweets.append(each.split())
#print(word_tweets)
str1=''
for each in tweets:
    str1 += each
word_from_tweets=(str1.split())
#print(word_from_tweets)
punct=string.punctuation
dig=string.digits
new_list=[]
new_list1=[]
stopwords=nltk.corpus.stopwords.words("english")
#extracted_words=[w for w in word_from_tweets not in stopwords]
#print(extracted_words)
for word in word_from_tweets:
    if word not in stopwords:
        remove=str.maketrans(punct,len(punct) * " ")
        w1=word.translate(remove)
        if len(w1.strip()) >3 and not w1.strip().isdigit():
            new_list.append(w1.strip()) 
#print(new_list)
freq= nltk.FreqDist(new_list)
freq.plot(25)
```
